```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd
```

```python
iris = load_iris()
type(iris)
```

```python
print (iris.data)
```

```python
print (iris.feature_names)
```

```python
x=iris.data
y=iris.target
#x-это признаки,a y-вид цветка
```

```python
selector= SelectKBest(score_func=chi2,k=2)
x_new= selector.fit_transform(x,y)
#используем selectkbest с функцией chi2 для того,чтобы получить лучшие признаки
```

```python
scores = selector.scores_  #оценки каждого признака
selected_features = selector.get_support(indices=True)  # индексы выбранных признаков

# оценки всех признаков
print("Оценки всех признаков:")
for i in range(len(iris.feature_names)):
    print(f"{iris.feature_names[i]}: {scores[i]:.1f}")

#Выводим названия выбранных признаков
print("\nТоп-2 выбранных признаков:")
for i in selected_features:
    print(f"{iris.feature_names[i]}: {scores[i]:.1f}")
```
